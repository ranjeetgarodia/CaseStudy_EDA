```python
## lets look for pct change as mean, values are having much variations across all country 
existing_df_pct_change = existing_df.pct_change()
existing_df_pct_change
```

```python
country_name=list(existing_df.columns.values)
```

```python
existing_df_pct_change[existing_df_pct_change[country_name] > 0.0].head()
```

```python
## total countries 207
## here number of countries whose % change increase w.r.t year more than 8 out of total 15 years data
tt=(existing_df_pct_change[existing_df_pct_change[country_name] > 0.03].count() >8)
print(tt.sum())
list(existing_df_pct_change.columns[tt])
```

### above countries are those whose %change >0.03 in existing cases
###1 have taken threshold value as 0.03 , nearly 10% of total country 
###2 have taken number of year count as 8 , atleast more than half of total year data 

```python
new_df_pct_change = new_df.pct_change()
new_df_pct_change.iloc[:,1]
```

```python
new_tt =new_df_pct_change[new_df_pct_change.iloc[7:,:] >0.03].count() >4
print('below are the countries for whom %change >0.03 in the recent years ::',new_tt.sum())
list(new_df_pct_change.columns[new_tt])
```
