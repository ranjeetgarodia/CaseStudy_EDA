```python
#What about world trends? 
#What is following code doing?
#print (deaths_df)

death_total_year = deaths_df.sum(axis =1) ## adding w.r.t to each row as axis=1
existing_case_total_year = existing_df.sum(axis =1)
new_case_total_year = new_df.sum(axis=1)

total_cases_year = pd.DataFrame({'death_total': death_total_year,
                   'existing_case_total': existing_case_total_year,
                   'new_case_total': new_case_total_year },
                   index = death_total_year.index)
total_cases_year
```

### plotting world trend eacn year cases

```python
total_cases_year.plot(figsize=(12,6)).legend(
   loc='best')
plt.show()
```

```python
death_country_mean = deaths_df.mean() ## will give mean for each country 
existing_country_mean = existing_df.mean()
new_cases_country_mean =new_df.mean()
```

```python
death_country_mean_summary = death_country_mean.describe()
death_country_mean_summary
```

```python
existing_country_mean_summary = existing_country_mean.describe()
existing_country_mean_summary
```

```python
new_cases_country_summary = new_cases_country_mean.describe()
new_cases_country_summary
```

```python
death_outlier = death_country_mean_summary['50%']*5
print('death_outlier value::',death_outlier)
outlier_countries_by_deaths_index=death_country_mean >death_outlier
print('number of outlier countries for death count::',sum(outlier_countries_by_deaths_index))

existing_outlier = existing_country_mean_summary['50%']*5
print('existing_outlier value::',existing_outlier)
outlier_countries_by_existingcaes_index=existing_country_mean >existing_outlier
print('number of outlier countries for existing cases count::',sum(outlier_countries_by_existingcaes_index))

new_cases_country_outlier = new_cases_country_summary['50%']*5
print('new_cases_country_outlier value::',new_cases_country_outlier)
outlier_countries_by_newcases_index=new_cases_country_mean >new_cases_country_outlier
print('number of outlier countries for new cases count::',sum(outlier_countries_by_newcases_index))
```
